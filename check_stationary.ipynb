```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
```

```python
np.random.seed(42)
random_walk = np.cumsum(np.random.randn(100))
```

```python
time_series = pd.Series(random_walk)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(time_series, label='Time Series')
plt.title('Time Series Data')
plt.legend()
plt.show()
```

```python
def adf_test(series):
    result = adfuller(series)
    print("Augmented Dickey-Fuller Test:")
    print(f"Test Statistic: {result[0]}")
    print(f"P-Value: {result[1]}")
    print("Critical Values:")
    for key, value in result[4].items():
        print(f"{key}: {value}")

    if result[1] < 0.05:
        print("Conclusion: The time series is stationary.")
    else:
        print("Conclusion: The time series is non-stationary.")
```

```python
adf_test(time_series)
```

```python
diff_series = time_series.diff().dropna()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(diff_series, label='Differenced Series')
plt.title('Differenced Time Series Data')
plt.legend()
plt.show()
```

```python
adf_test(diff_series)
```
